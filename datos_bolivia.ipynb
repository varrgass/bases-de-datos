```python
import matplotlib.pyplot as plt
from pytrends.request import TrendReq
import datetime

# Conectarse a Google
pytrend = TrendReq()

# Crear una lista de palabras clave
kw_list = ['TikTok']

# Especificar el rango de fechas (desde 2010 hasta hoy)
timeframe = f'2024-01-01 {datetime.datetime.now().strftime("%Y-%m-%d")}'

# Construir la carga útil
pytrend.build_payload(kw_list, timeframe=timeframe, geo='BO') # Usar 'BO' para Bolivia

# Interés por región
df = pytrend.interest_by_region()
# Remover la palabra "Department" de los nombres de departamentos
df.index = df.index.str.replace(" Department", "")
# Ordenar el dataframe por el interés en 'TikTok' facebook
df = df.sort_values(by='TikTok', ascending=False)

# Crear un gráfico de barras
plt.figure(figsize=(10,6))  # Ajustar el tamaño de la gráfica
df['TikTok'].plot(kind='bar')
plt.ylabel('Interés')
plt.xlabel('Departamentos')
plt.title('Interés en TikTok por departamento en Bolivia del 2024 hasta hoy')
plt.show()
```

```python
from pytrends.request import TrendReq
import pandas as pd

# Conectarse a Google
pytrend = TrendReq()

# Crear una lista de palabras clave para las aplicaciones y sitios de interés
kw_list = ['TikTok']

# Crear un DataFrame vacío para almacenar los resultados
df_total = pd.DataFrame()

# Recopilar los datos de los últimos 10 años
for year in range(2010, 2022):
    # Definir la línea de tiempo
    timeframe = f'{year}-01-01 {year+5}-12-31' if year <= 2016 else f'{year}-01-01 {datetime.datetime.now().strftime("%Y-%m-%d")}'

    # Construir la carga útil
    pytrend.build_payload(kw_list, timeframe=timeframe, geo='BO')  # geo='BO' se usa para Bolivia

    # Obtener interés a lo largo del tiempo
    df = pytrend.interest_over_time()

    # Obtener interés por región y agregar una columna 'dep' con el valor de la región
    df_region = pytrend.interest_by_region()
    df_region['dep'] = df_region.index

    # Agregar una columna 'geoName' con el valor 'Bolivia' a ambos dataframes
    df['geoName'] = 'Bolivia'
    df_region['geoName'] = 'Bolivia'

    # Agregar los dataframes a df_total
    df_total = df_total.append(df)
    df_total = df_total.append(df_region)

# Mostrar los resultados
print(df_total)
```
